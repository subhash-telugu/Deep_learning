```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
df=df.iloc[:,1:-2]
```

```python
df.head()
```

```python
df_label=df['diagnosis']
```

```python
df_indp=df.iloc[:,1:]
```

```python
df_std_indp=StandardScaler().fit_transform(df_indp)
```

```python
df_std_indp
```

```python
df_label_dep=LabelEncoder().fit_transform(df_label)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df_std_indp,df_label_dep,test_size=0.2)
```

```python
import torch
```

```python
x_train=torch.FloatTensor(x_train)
x_test=torch.FloatTensor(x_test)
y_train=torch.LongTensor(y_train)
y_test=torch.LongTensor(y_test)
```

```python

x_train.shape[1]
```

```python
import torch.nn as nn
class Model(nn.Module):
  def __init__(self,x_train):
    super().__init__()
    self.fc1=nn.Sequential(
        nn.Linear(x_train.shape[1],16),
        nn.ReLU()
    )
    self.fc2=nn.Sequential(
        nn.Linear(16,8),
        nn.ReLU()
    )
    self.fc3=nn.Sequential(
        nn.Linear(8,1),
        nn.Sigmoid()
    )
  def forward(self,features):
      x=self.fc1(features)
      x=self.fc2(x)
      x=self.fc3(x)
      return x
```

```python
model=Model(x_train)
```

```python


for name, param in model.named_parameters():
    if param.requires_grad:
        print(name, param.data)
```

```python
 loss=nn.BCELoss()
```

```python
optimizer=torch.optim.Adam(model.parameters(),lr=0.01)
```

```python
for epoch in range(100):
  y_pred=model(x_train)

  l=loss(y_pred,y_train.reshape(-1,1).float())

  optimizer.zero_grad()
  l.backward()
  optimizer.step()
  print(f'Epoch: {epoch + 1}, Loss: {l.item()}')

```

```python
# model evaluation
with torch.no_grad():
  y_pred = model.forward(x_test)
  y_pred = (y_pred > 0.7).float()
  accuracy = (y_pred == y_test).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
