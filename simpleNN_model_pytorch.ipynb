```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
df=df.iloc[:,1:-2]
```

```python
df_label=df['diagnosis']
```

```python
df_indp=df.iloc[:,1:]
```

```python
df_std_indp=StandardScaler().fit_transform(df_indp)
```

```python
df_std_indp
```

```python
df_label_dep=LabelEncoder().fit_transform(df_label)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df_std_indp,df_label_dep,test_size=0.2)
```

```python
import torch
```

```python
x_train=torch.FloatTensor(x_train)
x_test=torch.FloatTensor(x_test)
y_train=torch.LongTensor(y_train)
y_test=torch.LongTensor(y_test)
```

```python
x_train.shape[1]
```

```python
class SimpleNN:
  def __init__(self,x):
    self.weights=torch.rand(x.shape[1],1,requires_grad=True)
    self.bias=torch.rand(1,requires_grad=True)
  def forward(self,x):
    out=torch.matmul(x,self.weights)+self.bias
    out=torch.sigmoid(out)
    return out
  def loss_function(self, y_pred, y_train_tensor):
    # Clamp predictions to avoid log(0)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

    # Calculate loss
    loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
    return loss
```

```python
nn=SimpleNN(x_train)
```

```python
y_pred=nn.forward(x_train)
```

```python
nn.loss_function(y_pred,y_train)
```

```python
learning_rate=0.1
epochs=100
nn=SimpleNN(x_train)
for epoch in range(epochs):
  y_pred=nn.forward(x_train)
  loss=nn.loss_function(y_pred,y_train)
  loss.backward()
  with torch.no_grad():
    nn.weights-=learning_rate*nn.weights.grad
    nn.bias-=learning_rate*nn.bias.grad
  nn.weights.grad.zero_()
  nn.bias.grad.zero_()
  print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```
